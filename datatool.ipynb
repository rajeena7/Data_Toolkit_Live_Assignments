```python
'''14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and 
Perform the following tasks using Matplotlib:
 right y-axis for 'Humidity').
 b) Label the x-axis as 'Date'.
 a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and 
c) Set the title of the plot as 'Temperature and Humidity Over Time''''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a time-series dataset
dates = pd.date_range('2024-01-01', periods=100)
temperature = np.random.randint(20, 35, size=100)
humidity = np.random.randint(40, 70, size=100)

# Creating a DataFrame
data = {'Date': dates, 'Temperature': temperature, 'Humidity': humidity}
df = pd.DataFrame(data)

# Ploting Temperature and Humidity over time
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df['Date'], df['Temperature'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)  
ax2.plot(df['Date'], df['Humidity'], color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Temperature and Humidity Over Time')

fig.tight_layout()  
plt.show()
```

```python
'''15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following 
tasks using Matplotlib:
 a) Plot a histogram of the data with 30 bins.
 b) Overlay a line plot representing the normal distribution's probability density function (PDF).
 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 d) Set the title of the plot as 'Histogram with PDF Overlay'.'''

#16.Set the title of the plot as 'Histogram with PDF Overlay'.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Creating a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.5, color='skyblue', edgecolor='black', label='Histogram')

# Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show plot
plt.show()
```

```python
'''17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the 
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.'''
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generating two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determining the quadrants for each point
quadrants = np.zeros_like(x, dtype=int)
quadrants[(x >= 0) & (y >= 0)] = 1  # 1st Quadt
quadrants[(x < 0) & (y >= 0)] = 2   # 2st Quadt
quadrants[(x < 0) & (y < 0)] = 3    # 3st Quadt
quadrants[(x >= 0) & (y < 0)] = 4   # 4st Quadt

# Creating a DataFrame for the data
data = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrants})

# Creating a Seaborn scatter plot with color based on quadrants
sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette='viridis', legend='full')

# Add legend
plt.legend(title='Quadrants')

# Label axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set title
plt.title('Quadrant-wise Scatter Plot')

# Show plot
plt.show()
```

```python
'''18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine 
Wave Function'''
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Ensuring output
output_notebook()

# Generating data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Creating a new plot
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add a line renderer
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Customize grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the plot
show(p)
```

```python
'''19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their 
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical 
Bar Chart'''
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256

# Ensuring the output 
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a new plot
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', toolbar_location=None, tools="")

# Add a bar renderer with color mapping based on values
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize the plot
p.legend.title = "Categories"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

# Generate a random dataset
x = np.linspace(0, 10, 100)
y = np.random.randn(100).cumsum()

# Create a line plot
trace = go.Scatter(
    x = x,
    y = y,
    mode = 'lines',
    name = 'Random Data'
)

data = [trace]

layout = go.Layout(
    title = 'Simple Line Plot',
    xaxis = dict(title='X-axis'),
    yaxis = dict(title='Y-axis')
)

fig = go.Figure(data=data, layout=layout)

# Display the plot in the notebook
pyo.iplot(fig)
```

```python
'''21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set 
the title as 'Interactive Pie Chart''''
```

```python
import numpy as np
import plotly.graph_objs as go
import plotly.offline as pyo

# Generate random data
np.random.seed(42)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a pie chart
fig = go.Figure(
    data=[go.Pie(
        labels=categories,
        values=values,
        hoverinfo='label+percent',
        textinfo='value+percent',
        textfont_size=15
    )]
)

# Set the title
fig.update_layout(title_text='Interactive Pie Chart')

# Display the plot
pyo.iplot(fig)
```
