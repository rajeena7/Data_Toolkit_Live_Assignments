```python
import pandas as pd
import numpy as np
import re



```

```python
df=pd.read_csv('People Data.csv')
```

```python
df
```

```python
# Clean and transform the 'Phone' column
def clean_phone(phone):
    if pd.isna(phone):
        return -1  # Placeholder for missing values
    elif isinstance(phone, float):  # Check if float
        return int(phone)
    else:  # Assume string
        digits = re.findall(r'\d+', phone)
        if digits:
            return int(''.join(digits))
        else:
            return -1  # Placeholder for invalid phone numbers

df['Phone'] = df['Phone'].apply(clean_phone).astype(int)
```

```python
# Display table attributes and data types
print("Table Attributes:")
print(df.info())
```

```python
df.to_csv('cleaned_data.csv', index=True)
```
